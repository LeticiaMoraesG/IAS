```python
import collections.abc
collections.Mapping = collections.abc.Mapping

from pyknow import *
```

```python
class ClassificacaoGenerosMusicais(KnowledgeEngine):
    # Características Musicais Base
    @Rule(OR(
        Fact('instrumentação eletrônica'),
        Fact('sintetizador proeminente')
    ))
    def base_eletronica(self):
        self.declare(Fact('base_eletronica'))

    @Rule(OR(
        Fact('instrumentação acústica'),
        Fact('som orgânico')
    ))
    def base_acustica(self):
        self.declare(Fact('base_acustica'))

    # Fundações Rítmicas
    @Rule(OR(
        Fact('batida repetitiva'),
        Fact('orientado para dança')
    ))
    def base_ritmica(self):
        self.declare(Fact('base_ritmica'))

    @Rule(AND(
        Fact('base_eletronica'),
        OR(
            Fact('textura minimalista'),
            Fact('arranjo esparso')
        )
    ))
    def eletronica_minimal(self):
        self.declare(Fact('eletronica_minimal'))

    @Rule(AND(
        Fact('base_eletronica'),
        OR(
            Fact('atmosfera sombria'),
            Fact('elementos industriais')
        )
    ))
    def eletronica_escura(self):
        self.declare(Fact('eletronica_escura'))

    # Estilos com Base em Guitarra
    @Rule(OR(
        Fact('guitarras distorcidas'),
        Fact('ritmo rock')
    ))
    def base_guitarra(self):
        self.declare(Fact('base_guitarra'))

    @Rule(AND(
        Fact('base_guitarra'),
        OR(
            Fact('ritmo agressivo'),
            Fact('distorção pesada')
        )
    ))
    def caracteristicas_metal(self):
        self.declare(Fact('caracteristicas_metal'))

    @Rule(AND(
        Fact('base_guitarra'),
        OR(
            Fact('assinaturas rítmicas complexas'),
            Fact('letras filosóficas')
        )
    ))
    def caracteristicas_progressivo(self):
        self.declare(Fact('caracteristicas_progressivo'))

    # Complexidade Harmônica
    @Rule(OR(
        Fact('harmonia jazz'),
        Fact('improvisação')
    ))
    def complexidade_harmonica(self):
        self.declare(Fact('complexidade_harmonica'))

    @Rule(AND(
        Fact('complexidade_harmonica'),
        OR(
            Fact('influências música mundial'),
            Fact('elementos de fusão')
        )
    ))
    def caracteristicas_fusao(self):
        self.declare(Fact('caracteristicas_fusao'))

    # Classificações Específicas de Gênero
    @Rule(AND(
        Fact('eletronica_minimal'),
        Fact('base_ritmica'),
        Fact('bpm alto')
    ))
    def genero_techno(self):
        self.declare(Fact(genero='Techno'))

    @Rule(AND(
        Fact('eletronica_escura'),
        Fact('elementos industriais'),
        Fact('ritmo agressivo')
    ))
    def genero_industrial(self):
        self.declare(Fact(genero='Industrial'))

    @Rule(AND(
        Fact('caracteristicas_metal'),
        Fact('ritmo agressivo'),
        Fact('instrumentação técnica')
    ))
    def genero_metal_progressivo(self):
        self.declare(Fact(genero='Metal Progressivo'))

    @Rule(AND(
        Fact('base_guitarra'),
        Fact('caracteristicas_progressivo'),
        Fact('álbum conceitual')
    ))
    def genero_art_rock(self):
        self.declare(Fact(genero='Art Rock'))

    @Rule(AND(
        Fact('complexidade_harmonica'),
        Fact('caracteristicas_fusao'),
        Fact('manipulação eletrônica')
    ))
    def genero_fusao_global(self):
        self.declare(Fact(genero='Fusão Global'))

    @Rule(AND(
        Fact('base_eletronica'),
        Fact('base_acustica'),
        Fact('composição não convencional')
    ))
    def genero_experimental(self):
        self.declare(Fact(genero='Experimental'))

    @Rule(Fact(genero=MATCH.g))
    def imprimir_resultado(self, g):
        print(f'Gênero Musical Identificado: {g}')

    def factz(self, l):
        for x in l:
            self.declare(x)
```
